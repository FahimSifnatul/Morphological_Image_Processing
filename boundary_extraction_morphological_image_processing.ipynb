```python
# Boundary Extraction - morphological image processing
# learning resource - https://www.slideshare.net/maria_akther/boundary-extraction
# learning resource - https://www.cis.rit.edu/class/simg782.old/lec_morphology.html
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline

'''
(A ^ B) - binary image complement operation
(A & B) - binary image intersection
(A | B) - binary image union
(A & (B ^ 1)) - binary image set difference
'''
```

```python
# n x m image kernel / structuring element initialization
n, m = 5, 5 # kernal dimensions should be odd
kernel = np.ones((n, m), dtype = np.int32)
```

```python
plt.rcParams["figure.figsize"] = (10, 6)
RGB_image = mpimg.imread("image.png")
plt.imshow(RGB_image)
plt.title("Original Image")
```

```python
plt.rcParams["figure.figsize"] = (12, 10)
# RGB to binary conversion
binary_image = np.sum(RGB_image, axis = 2) / 3
binary_image = np.where(binary_image > 0.75, 0, 1)

# original binary image show
plt.subplot(2, 2, 1)
plt.imshow(binary_image, cmap = "binary")
plt.title("Original Binary Image")



# padded binary image
row, col = binary_image.shape
padded_binary_image = np.zeros((row + n - 1, col + m - 1), dtype = np.int32)
padded_binary_image[n//2 : -(n//2), m//2 : -(m//2)] = binary_image[:, :].copy()

# padded original binary image show
plt.subplot(2, 2, 2)
plt.imshow(padded_binary_image, cmap = "binary")
plt.title("Padded Binary Image")



# dilation - morphological image processing
row, col      = binary_image.shape
dilated_image = binary_image.copy()

for i in range(row):
    for j in range(col):
        tmp_img = padded_binary_image[i : i+n, j : j+m].copy()
        if np.sum(tmp_img == kernel) > 0:
            dilated_image[i, j] = 1
        else:
            dilated_image[i, j] = 0
            
# Boundary Extraction by dilation - morphological image processing
# boundary(A) = set_difference(dilation(A,kernel), A)
boundary_extracted_image_by_dilation = (dilated_image & (binary_image ^ 1))

# Boundary extracted image (by dilation) show
plt.subplot(2, 2, 3)
plt.imshow(boundary_extracted_image_by_dilation, cmap = "binary")
plt.title("Boundary extracted image (by dilation)")



# erosion - morphological image processing
row, col     = binary_image.shape
eroded_image = binary_image.copy()

for i in range(row):
    for j in range(col):
        tmp_img = padded_binary_image[i : i+n, j : j+m].copy()
        if np.sum(tmp_img == kernel) == n*m:
            eroded_image[i, j] = 1
        else:
            eroded_image[i, j] = 0

# Boundary Extraction by erosion - morphological image processing
# boundary(A) = set_difference(A, erosion(A,kernel))
boundary_extracted_image_by_erosion = (binary_image & (eroded_image ^ 1))
            
# Boundary extracted image (by erosion) show
plt.subplot(2, 2, 4)
plt.imshow(boundary_extracted_image_by_erosion, cmap = "binary")
plt.title("Boundary extracted image (by erosion)")
```
